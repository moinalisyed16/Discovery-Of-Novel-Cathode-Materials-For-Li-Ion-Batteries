### Checking Performance Of The Model 

```python
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
#Performance metrics
#In testing
r2_ML_y1_test = r2_score(y_test, y_pred_ML)
mae = mean_absolute_error(y_test, y_pred_ML)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_ML))

# Print the results
print('R2:', r2_ML_y1_test)
print('MAE:', mae)
print("RMSE:", rmse)
```

```python
#Normalizing the data

from sklearn.preprocessing import MinMaxScaler

# Create an instance of the MinMaxScaler class
scaler = MinMaxScaler()

# Normalize the data
data_std = scaler.fit_transform(data)
data_std=pd.DataFrame(data_std,columns=data.columns)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

### Applying SVM Regression On The Final Data

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

# Define the SVM regressor and hyperparameters to tune
svm = SVR()
hyperparams = {
    'kernel': ['linear', 'rbf'],
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'epsilon': [0.01, 0.1, 1]
}

# Use GridSearchCV to find the best hyperparameters
svm_grid = GridSearchCV(estimator=svm, param_grid=hyperparams, cv=5, scoring='neg_mean_squared_error')
svm_grid.fit(x_train, y_train)

best = svm_grid.best_params_
```

```python
# Create the SVR object
svr = SVR(kernel=best['kernel'], C=best['C'], epsilon=best['epsilon'])

# Train the model
svr.fit(x_train, y_train)

# Make predictions on the test data
y_pred_S = svr.predict(x_test)

```
